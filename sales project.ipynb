```python
data['Date']=pd.to_datetime(data['Date']) #Safe conversion
data['Month']=data['Date'].dt.month_name()

monthly_sales=data.groupby('Month')['Total'].sum().sort_values(ascending=False)
print(monthly_sales)
```

```python
data['Day']=data['Date'].dt.day_name()
top_day=data.groupby('Day')['Total'].sum().sort_values(ascending=False)
print(top_day)
```

```python
payment=data.groupby('Payment')['Total'].sum().sort_values(ascending=False)
print(payment)
```

```python
best_city=top_city.idxmax()
print("Best Performing City:",best_city)

best_product=top_product.idxmax()
print("Best Product Line",best_product)
```

```python
data.groupby('Day')['Total'].sum().plot(kind='bar',color='green',figsize=(7,4))
plt.title('Total sales by Day')
plt.ylabel('Total sales')
plt.xlabel('Day')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
